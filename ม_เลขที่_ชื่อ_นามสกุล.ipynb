# **5. ตัวดำเนินการทางคณิตศาสตร์**

<font size=4 color='blue'>**ตัวอย่างที่ 5.1** คำสั่งตัวดำเนินการทางคณิตศาสตร์</font>

```python
#ตัวอย่างที่ 5.1
a = 10                            #บรรทัดที่ 1 กำหนดค่าตัวแปร a เท่ากับ 10
b = 3                              #บรรทัดที่ 2 กำหนดค่าตัวแปร b เท่ากับ 3
print(a + b)                     #บรรทัดที่ 3 แสดงข้อมูล (a บวก b)
print(a - b)                      #บรรทัดที่ 4 แสดงข้อมูล (a ลบ b)
print(a * b)                     #บรรทัดที่ 5 แสดงข้อมูล (a คูณ b)
print(a / b)                     #บรรทัดที่ 6 แสดงข้อมูล (a หาร b)
print(a ** b)                   #บรรทัดที่ 8 แสดงข้อมูล (a ยกกำลัง b)
print(a // b)                    #บรรทัดที่ 9 แสดงข้อมูล (a หาร b เอาจำนวนเต็ม)
print(a % b)                   #บรรทัดที่ 10 แสดงข้อมูล (a หาร b เอาเศษที่เหลือ)
```

<font size=4 color='blue'>**ตัวอย่างที่ 5.2** เขียนคำสั่งไพทอน เพื่อคำนวณหาพื้นที่สี่เหลี่ยมผืนผ้า
โดยกำหนดให้ความกว้างเท่ากับ 5 และความยาวเท่ากับ 8</font>

```python
#ตัวอย่างที่ 5.2
wide = 5                                                #บรรทัดที่ 1 กำหนดค่าตัวแปรกว้างเป็น wide = 5
length = 8                                              #บรรทัดที่ 2 กำหนดค่าตัวแปรยาวเป็น length = 8
area = wide * length                                #บรรทัดที่ 3 กำหนดค่าตัวแปรพื้นที่เป็น พื้นที่ = กว้าง x ยาว
print ("ความกว้าง = " , wide)                      #บรรทัดที่ 4 แสดงข้อมูล ความกว้าง ตัวแปร wide
print ("ความยาว = " , length)                     #บรรทัดที่ 5 แสดงข้อมูล ความยาว ตัวแปร length
print ("พื้นที่สี่เหลี่ยมผืนผ้า = " , area)          #บรรทัดที่ 6 แสดงข้อมูล พื้นที่ ตัวแปร area
```
